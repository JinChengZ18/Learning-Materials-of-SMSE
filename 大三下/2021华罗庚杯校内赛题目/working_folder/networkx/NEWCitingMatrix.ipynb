```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 数据加载
data = pd.read_csv("NEWmatrix.csv",error_bad_lines=False) # 跳过导致混乱的行
data.head()
```

```python
data2 = pd.read_csv("NEWresult_all.csv",error_bad_lines=False)
data2.head()
```

```python
journal_lst = data2['journal_name']
target_lst  = data['Unnamed: 0']
Paps = data2['Total Articles']
Refs = data2['Number of References']
```

```python
orders = [np.where(journal_lst==target)[0][0] for target in target_lst]
```

```python
results = np.array(data.iloc[:,1:])
for i in range(results.shape[0]):
    for j in range(results.shape[1]):
        result = results[i][j]
        try:
            results[i][j] = result/Paps[orders[j]]*10**7
            results[i][j] = result/Refs[orders[i]]*10**7
        except:
            print('ERROR')
results = pd.DataFrame(results)
results.to_csv('NEWCitingMatrix.csv')
```
